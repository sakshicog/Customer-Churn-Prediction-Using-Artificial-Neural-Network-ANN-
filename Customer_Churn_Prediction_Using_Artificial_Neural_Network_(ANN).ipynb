```python
#splitting the dataset into two parts X and Y

dataX = pd.DataFrame(data.iloc[:,3:13].values)

dataY = data.iloc[:,13].values
```

```python
dataX.head()
```

```python
print(dataY)
```

```python
#Transforming the categorical data into numerical values
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

labelencoder_X_2 =LabelEncoder()
dataX.loc[:,2] = labelencoder_X_2.fit_transform(dataX.iloc[:,2])
```

```python
dataX.head()
```

```python

from sklearn.preprocessing import LabelEncoder, OneHotEncoder

labelencoder_X_1 =LabelEncoder()
dataX.loc[:,1] = labelencoder_X_1.fit_transform(dataX.iloc[:,1])
```

```python
from sklearn.compose import ColumnTransformer

#using columnTransformer to apply OnehotEncoder to the second column

# ColumnTransformer - It is  a tool allows to apply different preprocessimg transformation  to specfic columns of a dataset in a single step.

ct = ColumnTransformer(
    transformers=[
        ('onehot' , OneHotEncoder(drop= 'first'), [1])
    ],
    remainder='passthrough'
)

dataX = ct.fit_transform(dataX)
```

```python
dataX
```

```python
dataX.shape
```

Now spliting of the following processeed data into 80%(for training of data for creating a neural naertwiork model)
and 20%(for its test evaluation of the data)


```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dataX, dataY, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

Now Building Artificial Neural Network (ANN) using keras, tensorflow and pytorch libariaes

```python
import keras
from keras.models import Sequential
from keras.layers import Dense #The neural Network is fully connected
```

```python
classifier = Sequential()
classifier.add(Dense(6, kernel_initializer='uniform', activation='relu', input_dim=11))
classifier.add(Dense(6, kernel_initializer='uniform', activation='relu'))
classifier.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
```

Training Configuration
Compiole the ANN and fit the training data

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

classifier.fit(x_train, y_train, batch_size=10, epochs=50)
```

```python
y_pred= classifier.predict(x_test)
y_pred= (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
